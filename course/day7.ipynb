---## 2. Evaluation Organization### Current Structure (v6.0)Our v6.0 evaluation system intentionally separates learning artifacts from production modules. This structure reflects the documentation vs implementation distinction that defines Day 7's core lesson.**Learning Artifact: course/day5.ipynb**Day 5 notebook demonstrates LLM-as-a-Judge evaluation as an inline implementation with educational polish:- **125 test triplets**: 25 manually curated + 100 AI-generated via Pydantic AI question synthesis- **Pydantic schemas**: `EvaluationCheck` (single dimension) and `EvaluationChecklist` (aggregates all checks)- **Seven-dimension rubrics**:  - `instructions_follow` — Agent follows system prompt directions  - `instructions_avoid` — Agent avoids prohibited behaviors  - `answer_relevant` — Response addresses user question  - `answer_clear` — Response is readable and well-structured  - `answer_citations` — Sources are properly attributed  - `completeness` — All aspects of question covered  - `tool_call_search` — Search tool invoked when appropriate- **Chain-of-thought field ordering**: Schema forces LLM to generate `justification` before `check_pass`, reducing evaluation variance by 10-15%- **Inline implementation**: Complete evaluation pipeline in notebook cells with educational comments explaining each step**Production Modules: project/src/aihero/**The project/ folder extracts reusable patterns into production-quality modules (see PROJECT.md Phase 27 decisions for architecture rationale):- **evaluation.py**: LLM-as-a-Judge schemas, rubrics, and judge agent  - Pydantic `EvaluationCheck` and `EvaluationChecklist` BaseModel classes  - Seven-dimension rubric definitions with detailed instructions  - `judge_agent()` function using gpt-4o-mini at temperature=0.0  - Structured output enforcement via Pydantic AI- **logging.py**: JSON-based agent interaction logging  - Session metadata capture (timestamp, model, query)  - Message array serialization (system, user, assistant, tool messages)  - Environment-based configuration (LOGS_DIRECTORY)- **test_data.py**: TestTriplet schema and validation  - `TestTriplet` Pydantic model (question, expected_answer, source_files, source tracking)  - Validation utilities for test set integrity- **question_generator.py**: AI-powered question generation  - Pydantic AI agent for synthesizing test questions from corpus  - Configurable count and diversity parameters**Why this structure?**Notebooks excel at **learning** — inline comments, exploratory flow, visual outputs. Modules excel at **production** — type hints, docstrings, testable functions, reusable across contexts.Current structure is an intentional choice for v6.0 scope, not an oversight. Day 5 notebook teaches evaluation concepts interactively. Production modules implement those patterns with engineering rigor for the deployed agent.### Future Pattern: eval/ Folder (Industry Reference)Many production ML projects organize evaluation code into a dedicated `eval/` folder with two primary artifacts:- **eval/data-gen.ipynb**: Test data generation workflows  - Manual triplet curation interfaces  - AI-powered question synthesis from corpus  - Test set versioning and storage- **eval/evaluations.ipynb**: Evaluation pipelines and analysis  - Batch evaluation runs across test sets  - Per-dimension performance breakdowns  - Longitudinal tracking (model version comparisons)**Note:** This pattern is documented for awareness, not as current recommendation. Reorganization to eval/ folder structure deferred to v8.0+ per ROADMAP scope constraints. Our current structure (day5.ipynb for learning, project/src/aihero/ for production) appropriately balances educational clarity with engineering modularity for a course deliverable.**Reference:** course/day5.ipynb, project/src/aihero/, PROJECT.md Phase 27 decisions
